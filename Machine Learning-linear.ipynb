```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

# LInear Regression

```python
#linear can be done only on Numerical data
#Supervised Learning means we have to give the input and ouput
#automatic reshape
#dataset 
x=np.array([1,2,3,4,5,6,7]).reshape(-1,1)
print(x)
y=np.array([8,9,6,7,8,4,5])
print(y)
```

```python
#Model=LinearRegression(x,y)
Model=LinearRegression().fit(x,y)
print(Model)
```

```python
#find coefficients
#can be used as in FIT
#actual - the original is the error rate (x-y)
r_sq=Model.score(x,y)
print(r_sq)
print("intercept",Model.intercept_)
print("slope",Model.coef_)
```

```python
#pass to New Model
New_Model=LinearRegression().fit(x,y.reshape(-1,1))
print(New_Model)
print("intercept",New_Model.intercept_)
print("slope",New_Model.coef_)
```

```python
#Find the y Values 
y_pred=Model.predict(x)
print("Predicton Response",y_pred)
```

```python
#OR(y=b0+b1x)
y_pred1=Model.intercept_+Model.coef_*x
print("Predicton Response",y_pred1)
```

```python
#Neew input thats is New Model
x_new=np.arange(6).reshape(-1,1)
print(x_new)
y_new=Model.predict(x_new)
print(y_new)
```

```python
#Ploting the Values
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x_new,y_new)
plt.show()
```

```python
#Usiing Linespace
x1=np.linspace(-1,1,100)
print(x1)
y1=np.linspace(-1,2,100)
y1

plt.scatter(x1,y1)
plt.show()
```
